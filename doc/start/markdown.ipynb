## Langage de balisage

En informatique, un langage de balisage permet d'enrichir l'information textuelle. L'inclusion de balises permet de changer l'apparence du document.

Un langage de balisage est appelé **markup language** en anglais.  
Le texte contient des **balises** spéciales (=markup) pour formatter le texte.

### HTML
Le plus célèbre des langages à balises est **HTML**. C'est le langage dans lequel les pages internet sont écrites. D'ailleurs le **ML** dans HTML est pour **Hyper-Text Markup Langage**

Voici un exemple de code HTML utilisant la balise `<img>`

    <img src='https://upload.wikimedia.org/wikipedia/commons/thumb/4/48/Markdown-mark.svg/200px-Markdown-mark.svg.png'>
    
qui insère cette image dont l'originale se trouve sur Wikipédia.

<img src='https://upload.wikimedia.org/wikipedia/commons/thumb/4/48/Markdown-mark.svg/200px-Markdown-mark.svg.png'>

et un autre exemple utilisant la balise `<a>`

    <a href='https://fr.wikipedia.org/wiki/Hypertext_Markup_Language'>HTML sur Wikipédia</>

qui produit cet hyperlien

 <a href='https://fr.wikipedia.org/wiki/Hypertext_Markup_Language'>HTML sur Wikipédia</a>
    

### XML
C'est l'abréviation de **eXtendend Markup Langage**. Ce langage est utilisé par exemple pour le nouveau format des fichiers Microsoft Office:

- Word (.docx)
- Excel (.xlsx)
- PowerPoint (.pptx)

Ou le **X** dans l'extension est pour **Open XML**

### LaTeX
C'est un langage à balises pour représenter des formules mathématiques. Ce langage est très apprécié par les mathématiciens et scientifiques. Votre support de cours et vos exercices de maths ont certainement été écrits avec LaTeX.

Voici un exemple de code en LateX

    $$ c = \sqrt[2]{a^2 + b^2} $$
    
qui produit

$$ c = \sqrt{a^2 + b^2} $$

Souvent un ou deux symboles suffisent déjà pour la mise en format.

## Data visualization

### Subtask:
Visualize the power consumption and temperature data.


**Reasoning**:
Visualize the power consumption ('P_1_2') and temperature data for each house, and explore the relationships between 'P_1_2' and other features ('V_1_2', 'I_1_2').



```python
import matplotlib.pyplot as plt

# Convert 'DateHeure' to datetime if not already
if not pd.api.types.is_datetime64_any_dtype(combined_df['DateHeure']):
    combined_df['DateHeure'] = pd.to_datetime(combined_df['DateHeure'])

# 1 & 2. Time-series plots of 'P_1_2'
plt.figure(figsize=(15, 5))
for house in combined_df['House'].unique():
    house_df = combined_df[combined_df['House'] == house]
    plt.plot(house_df['DateHeure'], house_df['P_1_2'], label=f'House {house}')
plt.xlabel('Date and Time')
plt.ylabel('Power Consumption (P_1_2)')
plt.title('Power Consumption (P_1_2) over Time for Each House')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Time-series plots of 'Temperature'
plt.figure(figsize=(15, 5))
for house in combined_df['House'].unique():
    house_df = combined_df[combined_df['House'] == house]
    plt.plot(house_df['DateHeure'], house_df['Temperature'], label=f'House {house}')
plt.xlabel('Date and Time')
plt.ylabel('Temperature')
plt.title('Temperature over Time for Each House')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Relationships between 'P_1_2' and other features
for house in combined_df['House'].unique():
    house_df = combined_df[combined_df['House'] == house]
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.scatter(house_df['V_1_2'], house_df['P_1_2'], alpha=0.5)
    plt.xlabel('V_1_2')
    plt.ylabel('P_1_2')
    plt.title(f'House {house}: P_1_2 vs V_1_2')

    plt.subplot(1, 2, 2)
    plt.scatter(house_df['I_1_2'], house_df['P_1_2'], alpha=0.5)
    plt.xlabel('I_1_2')
    plt.ylabel('P_1_2')
    plt.title(f'House {house}: P_1_2 vs I_1_2')
    plt.tight_layout()
    plt.show()
```

## Data analysis

### Subtask:
Analyze the power consumption data to identify periods of high consumption, calculate average consumption per house, and investigate correlations with temperature and time of day.


**Reasoning**:
Identify periods of high power consumption, calculate average consumption, investigate correlations with temperature, and analyze the impact of time of day on power consumption.



```python
import matplotlib.pyplot as plt
import numpy as np

# 1. High Consumption Periods
plt.figure(figsize=(15, 10))
for house in combined_df['House'].unique():
    house_df = combined_df[combined_df['House'] == house]
    threshold = np.percentile(house_df['P_1_2'], 95)
    high_consumption = house_df[house_df['P_1_2'] > threshold]
    plt.plot(house_df['DateHeure'], house_df['P_1_2'], label=f'House {house}')
    plt.scatter(high_consumption['DateHeure'], high_consumption['P_1_2'], color='red', label=f'High Consumption (House {house})')
plt.xlabel('Date and Time')
plt.ylabel('Power Consumption (P_1_2)')
plt.title('Power Consumption with High Consumption Periods')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# 2. Average Power Consumption
average_consumption = combined_df.groupby('House')['P_1_2'].mean()
print("Average Power Consumption per House:\n", average_consumption)


# 3. Correlation Analysis
plt.figure(figsize=(10, 6))
for house in combined_df['House'].unique():
    house_df = combined_df[combined_df['House'] == house]
    correlation = house_df['P_1_2'].corr(house_df['Temperature'])
    print(f"Correlation between P_1_2 and Temperature for House {house}: {correlation}")
    plt.scatter(house_df['Temperature'], house_df['P_1_2'], label=f'House {house}')
plt.xlabel('Temperature')
plt.ylabel('Power Consumption (P_1_2)')
plt.title('Power Consumption vs. Temperature')
plt.legend()
plt.tight_layout()
plt.show()


# 4. Time-of-Day Analysis
combined_df['Hour'] = combined_df['DateHeure'].dt.hour
hourly_avg = combined_df.groupby(['House', 'Hour'])['P_1_2'].mean().reset_index()
plt.figure(figsize=(12, 6))
for house in combined_df['House'].unique():
    house_hourly = hourly_avg[hourly_avg['House'] == house]
    plt.plot(house_hourly['Hour'], house_hourly['P_1_2'], label=f'House {house}')
plt.xlabel('Hour of Day')
plt.ylabel('Average Power Consumption (P_1_2)')
plt.title('Average Hourly Power Consumption')
plt.legend()
plt.xticks(range(24))
plt.tight_layout()
plt.show()
```
